# Read Data

```python
data = []
with open("8vers10.txt", "r") as f:
    for line in f:
        data.append(float(line.strip()))
```

# Plot

```python
plt.figure()
plt.plot(data)
```

# Binary ?

```python
from scipy.signal import find_peaks
```

```python
peaks, _ = find_peaks(data)

plt.figure()
plt.plot(data)
plt.scatter(peaks, np.array(data)[peaks], c="r")
```

# Binary !

```python
b = np.where(np.array(data)[peaks] > 1.5, 1, 0)
b
```

# Lookup Table

```python
import re

d = d.replace("\n", " ")

tab = re.split(" same| flip", d)
```

```python
tab = [s.strip() for s in tab]
```

```python
tab = [s.replace(' ', '\t') for s in tab]
```

```python
tab = "\n".join(tab)
```

```python
import io
import pandas as pd
```

```python
df = pd.read_csv(io.StringIO(tab), sep="\t", dtype=str, header=None)
```
